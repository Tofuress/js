```
    JavaScript выполняется в одном потоке (однопоточен), но может работать асинхронно. 
    Синхронный или блокирующий код — это код, который выполняется строго последовательно. Следующая строка не начнет выполняться, пока не завершится предыдущая. 
    Асинхронный или неблокирующий код позволяет 'запускать' долгие операции и не дожидаться их завершения, а сразу переходить к выполнению следующего кода. Когда долгая операция завершится, мы получим уведомление.
```

```
    Call Stack (Стек вызовов) это простая структура данных 'последним вошел — первым вышел' (LIFO). Когда функция вызывается, она 'кладется' на верхушку стека. Когда она завершает работу (`return`), она 'снимается' со стека.
    Web APIs / Node.js APIs это набор мощных, но потенциально 'долгих' функций, которые предоставляет сама среда (браузер или Node.js). Они работают в отдельных потоках! Примеры: `setTimeout`, `fetch`, `fs.readFile`. Когда JS-движок встречает вызов такой функции, он не выполняет ее сам, а отдает эту задачу среде и продолжает выполнять следующий синхронный код.
    Callback Queue (Очередь колбэков) это простая структура данных 'первым вошел — первым вышел' (FIFO). Когда среда (Web API) завершает свою долгую операцию (таймер сработал, файл прочитан), она не возвращает результат сразу в JS, а кладет функцию-колбэк в эту очередь.
    Event Loop (цикл событий) управляет выполнением асинхронного кода в JavaScript. Он проверяет стек вызовов и очередь колбэков, перенося готовые колбэки из очереди в стек для выполнения.
```

```
    Колбэк — это функция, передаваемая в другую функцию как аргумент, которая будет вызвана позже. Колбэки бывают синхронные (вызываются сразу) и асинхронные (вызываются после завершения операции).
    Error-first Callback Convention это стандартное соглашение для асинхронных колбэков в Node.js – первый аргумент колбэка всегда зарезервирован для объекта ошибки. Если операция прошла успешно, `err` будет равен `null` или `undefined`. Если произошла ошибка, `err` будет содержать объект ошибки, а остальные аргументы (`data`) будут `undefined`. Это заставляет вас всегда проверять ошибки в первую очередь.
```

```
    Callback Hell (или 'Пирамида обреченных') — эта ситуация возникает, когда много асинхронных операций вложены друг в друга. Код становится нечитаемым, похожим на пирамиду, сложным для отладки и поддержки.
    Проблемы «Пирамиды»:
    Трудно читать: логика 'скачет' по диагонали, а не идет сверху вниз.
    Трудно отлаживать: где именно произошла ошибка? В какой из вложенных функций? 
    Трудно поддерживать: попробуйте добавить еще один шаг в середину этой пирамиды. 
    Повторение кода: логика обработки ошибок часто дублируется на каждом уровне вложенности
```

```
    Есть способы сделать код на колбэках чище, например, вынося каждую функцию на верхний уровень и передавая их по имени. Но это не решает фундаментальную проблему: инверсия управления (Inversion of Control).
```

## Задание 7

```
Напишите код, который:
1.	Загружает данные пользователя с id = 1
2.	Затем загружает его посты
3.	Для первого поста загружает комментарии
4.	Выводит все полученные данные
5.	Обрабатывает возможные ошибки на каждом этапе
```

````typescript
function loadUserData(userId, callback) {
    console.log(`Загрузка данных пользователя ${userId}...`);
    setTimeout(() => {
        callback(null, { id: userId, name: 'Иван Иванов' });
    }, 1000);

}
function loadUserPosts(userId, callback) {
    console.log(`Загрузка постов пользователя ${userId}...`);
    setTimeout(() => {
        callback(null, ['Пост 1', 'Пост 2', 'Пост 3']);
    }, 1500);
}

function loadPostComments(postId, callback) {
    console.log(`Загрузка комментариев к посту ${postId}...`);
    setTimeout(() => {
        callback(null, ['Комментарий 1', 'Комментарий 2']);
    }, 800);
}

loadUserData(1, (err, user) => {
  if (err) {
      return console.error("Ошибка загрузки пользователя", err);
  }
  console.log("Пользователь:", user);

  loadUserPosts(user.id, (err, posts) => {
    if (err) {
        return console.error("Ошибка загрузки постов", err);
    }
    console.log("Посты:", posts);

    if (posts.length > 0) {
      loadPostComments(posts[0], (err, comments) => {
        if (err) {
            return console.error("Ошибка загрузки комментариев", err);
        }
        console.log("Комментарии к первому посту:", comments);
      });
    }
  });
});
````

## Задание 8

```
Создайте таймер обратного отсчета с использованием колбэков
function countdown(seconds, onTick, onComplete) {
Реализуйте функцию, которая:
1.	Каждую секунду вызывает onTick с оставшимся временем
2.	По завершении времени вызывает onComplete
3.	Используйте setInterval для отсчета времени
}

Пример использования:
console.log('Запуск таймера!');
countdown(
    5, 
    (timeLeft) => console.log(`Осталось: ${timeLeft} сек.`),
    () => console.log('Время вышло!')
);
```

````typescript
function countdown(seconds, onTick, onComplete) {
    const interval = setInterval(() => {
        if (seconds > 0) {
            onTick(seconds);
            seconds -= 1;
        }
        else {
            onComplete;
            clearInterval(interval);
        }
    }, 1000)
}

console.log('Запуск таймера!');
countdown(
    5, 
    (timeLeft) => console.log(`Осталось: ${timeLeft} сек.`),
    () => console.log('Время вышло!')
);
````
