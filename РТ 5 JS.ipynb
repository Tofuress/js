```
Node.js поставляется с большим набором встроенных модулей, которые предоставляют доступ к основным возможностям системы. Их не нужно устанавливать через `npm`. Их можно просто подключить через `require`
Примеры встроенных модулей:
    • `fs` (File System): для работы с файлами и папками.
    • `path`: для работы с путями к файлам (независимо от ОС). 
    • `http`: для создания HTTP-серверов и клиентов (низкоуровневый). 
    • `os`: для получения информации об операционной системе. 
    • `crypto`: для криптографических операций. 
И многие другие.
Почти каждая функция в модуле `fs` имеет две версии: 
Асинхронная (предпочтительная): `fs.readFile(path, callback)` или `fs.promises.readFile(path)`.
Синхронная (блокирующая): `fs.readFileSync(path)`. 
В серверных приложениях всегда используйте асинхронные версии, чтобы не блокировать Event Loop!
Другие полезные функции `fs`:
    • `fs.appendFile(path, data)`: дописать данные в конец файла.
    • `fs.readdir(path)`: прочитать содержимое директории (получить массив имен файлов).
    • `fs.mkdir(path)`: создать новую директорию.
    • `fs.rm(path)`: удалить файл или директорию.
    • `fs.stat(path)`: получить информацию о файле (размер, дата создания).
```

````typescript
const fs = require('fs/promises');

async function writeToFile(fileName, text) {
    try {
        await fs.writeFile(fileName, text);
        console.log('файл записан');
    } catch (error) {
        console.error('ошибка записи:', error.message);
    }
}

writeToFile('test.txt', 'hello node.js!');
// файл записан
````

```
Вместе с Node.js устанавливается утилита под названием npm (Node Package Manager). npm — это две вещи: 
1. Огромный онлайн-репозиторий (библиотека) готовых модулей (пакетов), написанных другими разработчиками — npmjs.com. 
2. Инструмент командной строки для установки, обновления и управления этими пакетами в вашем проекте.
Каждый Node.js проект начинается с создания файла `package.json`. Это главный конфигурационный файл проекта. Он содержит: 
Название проекта, версию, автора… 
Самое главное: список всех внешних пакетов (зависимостей), которые нужны вашему проекту.
Npm install <имя_пакета> - эта команда делает три вещи: 
1. Скачивает указанный пакет из репозитория npm. 
2. Помещает его в папку `node_modules` в вашем проекте. 
3. Автоматически добавляет запись о пакете в ваш `package.json` в раздел `dependencies`.
`dependencies`: Пакеты, которые нужны для работы вашего приложения в продакшене (например, `express`, `prisma`, `chalk`). Устанавливаются командой `npm install `. 
`devDependencies`: Пакеты, которые нужны только для разработки и тестирования (например, `jest`, `nodemon`, `eslint`). Они не будут устанавливаться на боевом сервере. Устанавливаются с флагом `-D`: `npm install nodemon -D`.
Папка `node_modules` может содержать тысячи файлов и весить сотни мегабайт. Правило: Никогда не добавляйте папку `node_modules` в Git! Она должна быть первой строчкой в вашем файле `.gitignore`. Любой другой разработчик, скачав ваш проект, просто выполнит `npm install`, и `npm` сам скачает все нужные пакеты, глядя в `package-lock.json`.
`npm uninstall ` удаляет пакет из `node_modules` и `package.json`.
`npm update` пытается обновить все пакеты до последних версий, разрешенных в `package.json`.
Секция `scripts` в `package.json` Это очень удобное место для хранения часто используемых команд проекта. Они запускаются с помощью `npm run <имя_скрипта>`.
```

## Задание 7

````typescript
// app.js
const fs = require('fs/promises');
const path = require('path');

async function main() {
    const [, , command, ...args] = process.argv;

    try {
        switch (command) {

            case 'read': {
                const content = await fs.readFile(args[0], 'utf-8');
                console.log(content);
                break;
            }

            case 'write': {
                await fs.writeFile(args[0], args[1]);
                console.log('файл записан');
                break;
            }

            case 'copy': {
                await fs.copyFile(args[0], args[1]);
                console.log('файл скопирован');
                break;
            }

            case 'delete': {
                await fs.unlink(args[0]);
                console.log('файл удалён');
                break;
            }

            case 'info': {
                const stats = await fs.stat(args[0]);

                console.log({
                    size: stats.size,
                    created: stats.birthtime,
                    modified: stats.mtime,
                    extension: path.extname(args[0]),
                    directory: path.dirname(args[0]),
                    fileName: path.basename(args[0])
                });

                break;
            }

            case 'writejson': {
                const data = {
                    name: 'Sofia',
                    age: 18
                };

                await fs.writeFile(
                    args[0],
                    JSON.stringify(data, null, 2)
                );

                console.log('json записан');
                break;
            }

            case 'readjson': {
                const json = await fs.readFile(args[0], 'utf-8');

                console.log(JSON.parse(json));

                break;
            }

            case 'help': {
                console.log(`
доступные команды:

read <file>
write <file> <text>
copy <source> <destination>
delete <file>
info <file>
writejson <file>
readjson <file>
`);
                break;
            }

            default:
                console.log('неизвестная команда');
        }

    } catch (error) {
        console.error('ошибка:', error.message);
    }
}

main();
````

````typescript
node app.js read test.txt
node app.js write test.txt Hello // файл записан
node app.js copy test.txt copy.txt // файл скопирован
node app.js delete copy.txt // файл удалён
node app.js info test.txt
node app.js writejson data.json // json записан
node app.js readjson data.json
node app.js help
````
