```
Людям, знакомым с другими языками программирования, такими, как Python и C/C++, синтаксис JS может показаться в целом знакомым. Например:
```

````typescript
Python:

a = 13
if a > 20:
    print("a is greater than 20")
elif a < 10:
    print("a is less than 10")
else:
    print("a is between 10 and 20)
````

```
    Переменные – так или иначе указатели или ссылки на данные, требуемые для работы программы, однако, «под капотом» их работа может значительно различаться.
    В статически типизируемых языках (Си) существует четкая связь между переменной и местом в памяти, где содержатся соответствующие ей данные.
    В динамически типизируемых языках, переменная содержит именно ссылку на место в памяти, при этом, эта ссылка может меняться, на что указывает оператор присваивания «=».
    Прежде, чем использовать переменные, их необходимо объявить:

```

```
        •	неявный let, когда указывается только название переменной

        •	объявление через var
    В большинстве случаев, переменные объявляются через явный let, а там, где это возможно – через const, что обеспечивает защиту от случайного изменения в дальнейшем.
    Переменные существуют в определенном «контексте», который переключается при помощи блоков, начало и конец которых обозначается фигурными скобками («{}»). 
```

## Задание 12.

```
Создайте функцию - калькулятор. Функция должна принимать три параметра: два числа и строку с операцией ('+', '-', '*', '/'). В зависимости от операции функция должна вернуть результат вычисления. Вызовите функцию.
```

````typescript
function calc(a, b, op) {
    switch (op) {
        case '+': return a + b;
        case '-': return a - b;
        case '*': return a * b;
        case '/': return b !== 0 ? a / b : "деление на 0";
        default: return "Ошибка";
    }
}


let res = calc(20, 5, '/');
console.log(res);
````

```

    Свойства-данные, соответственно, хранят в себе данные. Чаще всего это примитивы.
    Свойства-акксессоры позволяют создавать логическую «прослойку» между «сырыми» данными внутри объекта и их получателем, аналогичны get-терам и set-терам в Python-е
```

## Задание 13.

```
Создать функцию, которая, получая на вход два объекта, меняет местами их свойства: свойства объекта 1 становятся свойствами объекта 2 и наоборот.
    •	Так как объекты изменяемы, в том числе внутри функции, в которую они попали, к самим объектам не требуется применять оператор присваивания и функция не обязательно должна иметь вывод (return)
    •	Существуют неперечислимые свойства объектов, недоступные через интерфейс for…in
```

````typescript
function f(var1, var2) {
    let s = {};

    for (let i in var1) {
        s[i] = var1[i];
        delete var1[i];
    }

    for (let i in var2) {
        var1[i] = var2[i];
    }

    for (let i in s) {
        var2[i] = s[i];
    }
}

let a = { x: 1, y: 2 };
let b = { z: "3", k: "4" };

f(a, b);

console.log(a);
console.log(b);
````
