```
    Массив в JavaScript — это специализированный объект, предназначенный для хранения упорядоченных коллекций данных. Массивы представляют собой структуры данных, в которых каждый элемент ассоциирован с числовым индексом, начинающимся с нуля. Несмотря на то, что массивы являются объектами, они обладают внутренней оптимизацией для работы с последовательностями данных. Гетерогенность содержимого: Массивы могут содержать элементы различных типов данных одновременно, включая примитивные типы (числа, строки, булевы значения), объекты, другие массивы (создавая многомерные структуры), функции, а также null и undefined. В отличие от статически типизированных языков, массивы JavaScript автоматически изменяют свой размер при добавлении или удалении элементов, не требуя явного выделения или освобождения памяти.
```

```
    Методы массивов разделяются на три фундаментальные категории по их воздействию на исходный массив:
Мутирующие (изменяющие) методы — трансформируют исходный массив, изменяя его содержимое или структуру. Принципиальная характеристика: модификация происходит на месте (in-place) без создания нового массива.
    Немутирующие (неизменяющие) методы — возвращают новое значение или новый массив, оставляя исходный массив неизменным. Эти методы следуют принципам иммутабельности.
    Итерационные методы высшего порядка — принимают функции обратного вызова (callback) в качестве аргументов, применяя их к элементам массива. Могут быть как мутирующими, так и немутирующими.
    Метод forEach — это метод массивов, предназначенный для выполнения итерации по элементам массива. Он является типичным примером функции высшего порядка, так как принимает в качестве аргумента callback-функцию, которая будет выполнена для каждого элемента массива. Метод forEach последовательно перебирает элементы массива в порядке возрастания индексов и для каждого элемента вызывает переданную callback-функцию. Важно отметить, что forEach не создает новый массив, а просто выполняет заданные операции.
    Метод map()  — это метод преобразования (трансформации) массива, который создаёт новый массив, применяя заданную функцию к каждому элементу исходного массива. Новый массив той же длины, что и исходный. Каждый элемент нового массива — это результат вызова callback-функции для соответствующего элемента исходного массива.
    Метод filter() создает новый массив, в который попадают только те элементы исходного массива, для которых переданная функция-фильтр вернула true. Возвращает новый массив, который может быть короче исходного (или пустым).
    Метод reduce() превращает (сворачивает) весь массив в одно значение, последовательно обрабатывая каждый элемент. Берётся начальное значение (аккумулятор) — его можно задать самому или использовать первый элемент массива. По очереди для каждого элемента массива выполняется функция, которая принимает текущее значение аккумулятора и текущий элемент, и возвращает новое значение аккумулятора. После обработки всех элементов возвращается итоговое значение аккумулятора.
```

## Задание 14

```
1. Найдите средний возраст пользователей
2. Получите массив всех уникальных навыков
3. Создайте новый массив объектов, где каждый объект содержит имя пользователя и количество его навыков
4. Найдите пользователя, у которого больше всего навыков
```

````typescript
const users = [
    { id: 1, name: 'Алексей', age: 25, skills: ['JavaScript', 'React'] },
    { id: 2, name: 'Мария', age: 30, skills: ['Python', 'Django', 'HTML'] },
    { id: 3, name: 'Иван', age: 22, skills: ['JavaScript', 'CSS'] },
    { id: 4, name: 'Ольга', age: 28, skills: ['Java', 'Spring'] },
];

const avg = users.reduce((sum, user) => sum + user.age, 0) / users.length;
console.log('средний возраст:', avg);

const user_skills = [];
users.forEach(user => {
    user.skills.forEach(skill => {
        let found = false;
        for (let i = 0; i < user_skills.length; i++) {
            if (user_skills[i] === skill) {
                found = true;
                break;
            }
        }
        if (!found) {
            user_skills.push(skill);
        }
    });
});
console.log('уникальные навыки:', user_skills);

const skills_count = users.map(user => ({name: user.name, skills_count: user.skills.length}));
console.log('пользователи и количество навыков:', skills_count);

const best_user = users.reduce((max, user) => user.skills.length > max.skills.length ? user : max);
console.log('лучший пользователь с наибольшим кол-вом навыков:', best_user);
````

## Задание 15

```
1. Рассчитайте общую стоимость товаров в корзине без учета скидок
2. Примените скидки последовательно (первая скидка от общей суммы, вторая от уже уменьшенной)
3. Создайте новый объект заказа со структурой: { user: 'имя', total: итого, itemsCount: количество позиций }
4. Преобразуйте объект заказа в JSON строку
```

````typescript
const cart = {
    user: 'Петр',
    items: [
        { productId: 101, name: 'Молоко', price: 80, quantity: 2 },
        { productId: 205, name: 'Хлеб', price: 50, quantity: 1 },
        { productId: 312, name: 'Сыр', price: 300, quantity: 1 },
    ],
    discounts: [10, 5]
};

const total1 = cart.items.reduce((sum, item) => sum + (item.price * item.quantity), 0);
console.log('сумма без скидок:', total1);

let total2 = total1;
for (const i of cart.discounts) {
    total2 = total2 * (1 - i / 100);
}
console.log('сумма со скидками:', total2);

const order = {
    user: cart.user,
    total: total2,
    itemsCount: cart.items.reduce((count, item) => count + item.quantity, 0)
};
console.log('заказ:', order);

const json = JSON.stringify(order);
console.log('json заказа:', json);
````

## Задание 16

```
1. Разбейте текст на слова (по пробелам)
2. Создайте объект, где ключи - это слова, а значения - количество их повторений в тексте
3. Найдите самое длинное слово в тексте
4. Отсортируйте слова по частоте использования (от наиболее частых к наименее частым)
5. Преобразуйте итоговый объект статистики в JSON строку
```

````typescript
const text = "JavaScript это мощный язык программирования который используется для создания интерактивных веб страниц";

const words = text.split(' ');
console.log('слова:', words);

const words_rep = new Map();
words.forEach(word => {
    if (words_rep[word]) {
        words_rep[word] += 1;
    }
    else {
        words_rep[word] = 1;
    }
})
console.log('кол-во повторений:', words_rep);

let longest_word = words[0];
words.forEach(word => {
    if (word.length > longest_word.length) {
        longest_word = word;
    }
});
console.log('самое длинное слово:', longest_word);
console.log('длина:', longest_word.length);

const sort_array = [];
for (let word in words_rep) {
    sort_array.push({word: word, count: words_rep[word]});
}

sort_array.sort((a, b) => b.count - a.count);

console.log('отсортированные слова:', sort_array);

const json = JSON.stringify(words_rep, null, 2);
console.log('json', json);
````

## Задание 17

```
1. Получите массив доступных книг (available: true)
2. Найдите самую старую книгу в библиотеке
3. Сгруппируйте книги по автору 
4. Добавьте новую книгу в библиотеку
5. Экспортируйте данные библиотеки в JSON формат для передачи на сервер
```

````typescript
const library = {
    name: 'Центральная библиотека',
    books: [
        { id: 1, title: '1984', author: 'Джордж Оруэлл', year: 1949, available: true },
        { id: 2, title: 'Мастер и Маргарита', author: 'Михаил Булгаков', year: 1967, available: false },
        { id: 3, title: 'Преступление и наказание', author: 'Федор Достоевский', year: 1866, available: true },
        { id: 4, title: 'Гарри Поттер', author: 'Джоан Роулинг', year: 1997, available: true },
    ]
};

const available_books = [];
library.books.forEach(book => {
    if (book.available === true) {
        available_books.push(book);
    }
});
console.log('книги в наличии:', available_books);

const sorted = [...library.books];
sorted.sort((a, b) => a.year - b.year);
const oldest_book = sorted[0];
console.log('самая старая книга:', oldest_book);
console.log('название самой старой книги:', oldest_book.title);
console.log('год издания:', oldest_book.year);

const books_author = {};
library.books.forEach(book => {
    if (!books_author[book.author]) {
        books_author[book.author] = [];
    }
    books_author[book.author].push(book.title);
});
console.log('книги по авторам:', books_author);

const new_book = { 
    id: 5, 
    title: 'Война и мир', 
    author: 'Лев Толстой', 
    year: 1869, 
    available: true 
};
library.books.push(new_book);
console.log('после добавления:', library.books);

const json = JSON.stringify(library);
console.log(json);
````
