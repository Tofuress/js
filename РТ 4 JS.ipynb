```
    Промис - это объект, представляющий результат асинхронной операции. Любой промис в любой момент времени находится в одном из трех состояний: 
    1. `pending` (ожидание): Начальное состояние. Операция еще не завершилась.
    2. `fulfilled` (выполнено успешно): Операция завершилась успешно. Промис 'выполнился' с каким-то результатом. 
    3. `rejected` (отклонено): Операция завершилась с ошибкой. Промис 'отклонился' с какой-то ошибкой.
    Мы не можем получить результат напрямую. Вместо этого мы 'подписываемся' на завершение промиса с помощью специальных методов: `.then()`, `.catch()` и `.finally()`. Эти методы регистрируют колбэки, которые будут помещены в Microtask Queue.
    Macrotask Queue (или Task Queue): Очередь, куда попадают колбэки от `setTimeout`, кликов мыши, I/O-операций. (Наша старая 'Callback Queue'). 
    Microtask Queue: Специальная очередь с более высоким приоритетом. Сюда попадают колбэки от Промисов (`.then`, `.catch`).
    Правило Event Loop: После каждой Macrotask он выполняет ВСЕ задачи из Microtask Queue, и только потом берет следующую Macrotask.
    1. `.then(onFulfilled, onRejected)` принимает до двух аргументов-функций: 
    `onFulfilled`: выполнится, если промис успешно завершился. Примет результат в качестве аргумента.
    `onRejected`: выполнится, если промис был отклонен. Примет ошибку в качестве аргумента. (Используется редко).
    2. `.catch(onRejected)` — это синтаксический сахар для `.then(null, onRejected)`. Гораздо удобнее для обработки ошибок.
    3. `.finally(onFinally)` выполнится в любом случае, и при успехе, и при ошибке. Удобно для 'очистки' (например, убрать индикатор загрузки).
    Каждый вызов `.then()` (и `.catch()`, и `.finally()`) сам возвращает новый промис. Это позволяет выстраивать их в плоские, читаемые цепочки.
    Основными преимуществами цепочек промисов являются читаемость, единая обработка ошибок в конце цепочки и компонуемость(возможность легко добавлять, удалять или менять местами шаг `.then в цепочке).
    Промисы — это фундаментальный 'строительный блок' для современной асинхронности в JS. Они решили главные проблемы колбэков: 'пирамиду' и 'инверсию управления'.
```

## Задание 1.

```
Создайте промис, который будет отклонен (reject) через 2 секунды с ошибкой "Что-то пошло не так". Обработайте ошибку с помощью .catch() и выведите сообщение в консоль.Выводит "Работа завершена"
```

````typescript
function delay() {
  return new Promise((resolve, reject) => {
    setTimeout(() => {
      reject("Что-то пошло не так");
    }, 2000);
  });
}

const myPromise = delay();

myPromise.catch((error) => {
  console.log("Ошибка:", error);
});
````

```
    `async/await` — это не новая технология. Это 'синтаксический сахар' поверх Промисов. Он позволяет нам работать с промисами так, как будто это обычный, синхронный, блокирующий код, но без реальной блокировки Event Loop.
    Слово `async` ставится перед объявлением функции: `async function myFunction() { ... }` или `const myFunction = async () => { ... }`. Оно делает две вещи: 
    1. Позволяет использовать ключевое слово `await` внутри этой функции. 
    2. Гарантирует, что эта функция всегда вернет Промис.
    Оператор `await` ставится перед выражением, которое возвращает Промис. Он приостанавливает выполнение `async`-функции, 'дожидается', пока промис выполнится, и 'распаковывает' его результат. Важное правило: `await` можно использовать только внутри `async`-функции.
    Когда движок JS встречает `await`, он: 'Ставит на паузу' выполнение функции и возвращает управление в Event Loop, чтобы тот мог выполнять другие задачи. Когда промис разрешается, Event Loop 'возобновляет' выполнение функции с того места, где она остановилась.
    Ошибки в асинхронных функциях можно обрабатывать с помощью try...catch. Это делает код более читаемым по сравнению с .catch() в цепочках промисов.
```

## Задание 3

```
Напишите асинхронную функцию countdown, которая принимает число n и выводит числа от n до 1 с интервалом в 1 секунду, а затем выводит "Пуск!". Используйте await и функцию wait из примера.
```

````typescript
function wait(ms) {
  return new Promise((resolve) => setTimeout(resolve, ms));
}

async function countdown(n) {
  for (let i = n; i >= 1; i--) {
    console.log(i);
    await wait(1000);
  }
  console.log("Пуск!");
}

countdown(5);
````

```
Создайте асинхронную функцию getData, которая имитирует запрос к серверу. Используйте riskyOperation из примера. Если запрос успешен, выведите результат, иначе выведите сообщение об ошибке. Используйте try...catch.
```

````typescript
function riskyOperation() {
  return new Promise((resolve, reject) => {
    const random = Math.random();
    setTimeout(() => {
      if (random > 0.5) {
        resolve("Данные успешно получены");
      } else {
        reject("Ошибка сервера");
      }
    }, 1000);
  });
}

async function getData() {
  try {
    const result = await riskyOperation();
    console.log("Результат:", result);
  } catch (error) {
    console.log("Ошибка:", error);
  }
}

getData();
````

```
    Promise.all позволяет запустить несколько промисов параллельно и дождаться их выполнения. Этот новый промис выполнится (fulfill) тогда, когда ВСЕ промисы в массиве успешно выполнятся. Результатом будет массив их результатов. Он будет отклонен (reject), как только ХОТЯ БЫ ОДИН из промисов в массиве будет отклонен. Ошибкой будет ошибка этого первого упавшего промиса.
    Помимо Promise.all, существуют Promise.allSettled, Promise.race, Promise.any для различных сценариев параллельного выполнения.
    Promise.allSettled ждет, пока все промисы завершатся (неважно, с успехом или ошибкой). В результате выводит массив объектов с состоянием каждого промиса (`{ status: 'fulfilled', value: ... }` или `{ status: 'rejected', reason: ... }`). Стоит использовать когда вам нужно выполнить несколько независимых операций, и провал одной не должен отменять остальные.
    Promise.race ждет, пока самый первый промис завершится (успешно или с ошибкой), и принимает его результат/ошибку. Выводит результат или ошибку самого быстрого промиса. Стоит использовать когда у вас есть несколько источников данных, и вам достаточно ответа от самого быстрого. Или для реализации 'таймаута' для другой операции.
    Promise.any ждет, пока самый первый промис завершится УСПЕШНО. Выводит результат первого успешно выполненного промиса. В случае если все промисы отклонятся, то `Promise.any` сам отклонится с ошибкой `AggregateError`. Стоит использовать когда у вас есть несколько 'зеркал' одного ресурса, и вам достаточно получить ответ от любого из них.
```

## Задание 5

```
Создайте массив из 5 промисов, каждый из которых ждет случайное время от 1 до 3 секунд (используйте Math.random()). Используйте Promise.all, чтобы дождаться их выполнения и вывести массив времен, которые они ждали.
```

````typescript
function randomDelayPromise(index) {
  const delay = Math.floor(Math.random() * 2000) + 1000;
  return new Promise((resolve) => {
    setTimeout(() => {
      console.log(`Промис ${index} завершился через ${delay} мс`);
      resolve(delay);
    }, delay);
  });
}

const promises = [];
for (let i = 1; i <= 5; i++) {
  promises.push(randomDelayPromise(i));
}

Promise.all(promises).then((results) => {
  console.log("Все времена задержек:", results);
});
````

````typescript
const promise1 = new Promise((resolve) => {
  setTimeout(() => resolve("Первый"), 1000);
});

const promise2 = new Promise((resolve, reject) => {
  setTimeout(() => reject("Второй не удался"), 2000);
});

Promise.race([promise1, promise2])
  .then((result) => console.log("race - успех:", result))
  .catch((error) => console.log("race - ошибка:", error));

Promise.allSettled([promise1, promise2]).then((results) => {
  console.log("allSettled - результаты:");
  results.forEach((result, index) => {
    if (result.status === "fulfilled") {
      console.log(`Промис ${index + 1}: успех - ${result.value}`);
    } else {
      console.log(`Промис ${index + 1}: ошибка - ${result.reason}`);
    }
  });
});
````

````typescript
function wait(ms) {
  return new Promise((resolve) => setTimeout(resolve, ms));
}

function fetchUser() {
  return wait(1000).then(() => ({ id: 1, name: "Иван" }));
}

function fetchUserOrders(userId) {
  return wait(1500).then(() => [
    { orderId: 1, total: 1000 },
    { orderId: 2, total: 2000 },
  ]);
}

function fetchOrderDetails(orderId) {
  return wait(800).then(() => ({ orderId, items: ["Товар1", "Товар2"] }));
}
````

````typescript
async function parallelVersion() {
  const start = Date.now();
  const user = await fetchUser();
  const orders = await fetchUserOrders(user.id);

  const detailsPromises = orders.map((order) =>
    fetchOrderDetails(order.orderId)
  );
  const detailsArray = await Promise.all(detailsPromises);
  const ordersWithDetails = orders.map((order, index) => ({
    ...order,
    items: detailsArray[index].items,
  }));

  console.log(`Пользователь: ${user.name}`);
  console.log("Заказы:");
  ordersWithDetails.forEach((order) => {
    console.log(
      `  Заказ #${order.orderId} на сумму ${order.total} руб. (Товары: ${order.items.join(", ")})`
    );
  });

  const end = Date.now();
  console.log(`Общее время: ${end - start} мс`);
}

parallelVersion();
````

```
Создайте функцию withTimeout, которая принимает промис и время в миллисекундах и возвращает промис, который:
- разрешается с результатом исходного промиса, если он успевает выполниться за отведенное время
- отклоняется с ошибкой "Таймаут", если исходный промис не успевает
Используйте Promise.race для реализации таймаута.
Протестируйте withTimeout на быстром промиссе (500 мс) и медленном (2000 мс) с таймаутом 1000 мс.
```

````typescript
function withTimeout(promise, timeoutMs) {
  const timeoutPromise = new Promise((resolve, reject) => {
    setTimeout(() => {
      reject("Таймаут");
    }, timeoutMs);
  });

  return Promise.race([promise, timeoutPromise]);
}

const fastPromise = new Promise((resolve) => {
  setTimeout(() => resolve("Быстрый успех"), 500);
});

const slowPromise = new Promise((resolve) => {
  setTimeout(() => resolve("Медленный успех"), 2000);
});

withTimeout(fastPromise, 1000)
  .then((result) => console.log("Быстрый промис:", result))
  .catch((error) => console.log("Быстрый промис ошибка:", error));

withTimeout(slowPromise, 1000)
  .then((result) => console.log("Медленный промис:", result))
  .catch((error) => console.log("Медленный промис ошибка:", error));
````

## Задание 10

```
У вас есть массив URL-адресов (имитируйте их промисами, которые ждут случайное время). Напишите две функции:
1. sequentialRequests(urls) - делает запросы последовательно и возвращает массив результатов.
2. parallelRequests(urls) - делает запросы параллельно и возвращает массив результатов.
Измерьте время выполнения обеих функций и сравните. Обработайте ошибки так, чтобы при ошибке в одном запросе (например, отклонить промис для одного url) функция продолжала выполняться.

const urls = [
    'url1',
    'url2',
    'url3',
'url4'];

function mockRequest(url) {
    const delay = Math.floor(Math.random() * 2000) + 500; // Случайная задержка от 500 до 2500 мс
    return wait(delay).then(() => `Ответ от ${url} (задержка ${delay}мс)`);}
```

````typescript
function wait(ms) {
  return new Promise((resolve) => setTimeout(resolve, ms));
}

function mockRequest(url) {
  const delay = Math.floor(Math.random() * 2000) + 500;
  return wait(delay).then(() => `Ответ от ${url} (задержка ${delay}мс)`);
}

const urls = ["url1", "url2", "url3", "url4"];

async function sequentialRequests(urls) {
  const results = [];
  for (const url of urls) {
    try {
      const result = await mockRequest(url);
      results.push(result);
    } catch (error) {
      results.push(`Ошибка при ${url}: ${error.message}`);
    }
  }
  return results;
}

async function parallelRequests(urls) {
  const promises = urls.map((url) =>
    mockRequest(url).catch((error) => `Ошибка при ${url}: ${error.message}`)
  );
  return Promise.all(promises);
}

async function compare() {
  console.log("Последовательные запросы");
  let start = Date.now();
  const seqResults = await sequentialRequests(urls);
  console.log("Результаты:", seqResults);
  console.log(`Время: ${Date.now() - start} мс\n`);

  console.log("Параллельные запросы");
  start = Date.now();
  const parResults = await parallelRequests(urls);
  console.log("Результаты:", parResults);
  console.log(`Время: ${Date.now() - start} мс`);
}

compare();
````
